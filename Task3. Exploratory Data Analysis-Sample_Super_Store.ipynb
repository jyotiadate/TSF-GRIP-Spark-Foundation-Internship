```python
store=data.drop(['Segment'],axis=1)
store.head()
```

### Relationshop Analysis

```python
correlation=store.corr()
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True)
```

```python
sns.pairplot(store)
```

```python
sns.relplot(x='Category',y='Sub-Category',hue='Region',data=store)
```

```python
sns.relplot(x='Sales',y='Profit',hue='Category',data=store)
```

```python
sns.distplot(store['Sales'],bins=15)
```

```python
sns.catplot(x='Profit',kind='box',data=store)
```
